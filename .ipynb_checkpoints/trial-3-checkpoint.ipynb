```python
import mysql.connector
import pandas as pd

# connect to MySQL database
db = mysql.connector.connect(
    host="localhost",
    user="root",
    password="root",
    database="sys"
)

# define MySQL query
query = "SELECT employee_id, attendance_log_id, attendance_date, punch_records FROM stg_attendancelogs"

# execute MySQL query and read results into pandas DataFrame
df = pd.read_sql_query(query, db)

# create empty lists for punch in and out times
punch_in_times = []
punch_out_times = []
attendance_dates = []

# loop through punch records and extract punch in and out times and attendance dates
for attendance_date, punch in zip(df['attendance_date'], df['punch_records']):
    # split punch record into individual events
    events = punch.split(',')
    
    # initialize variables for tracking punch in and out times
    punch_in_time = None
    punch_out_time = None
    
    # loop through events and extract punch in and out times
    for e in events:
        # split event into time and event/direction
        split_e = e.split(':')
        if len(split_e) < 2:
            continue
        # extract time and event information
        time = ':'.join(split_e[:2])
        event_direction = split_e[2]
        if '(' in event_direction and ')' in event_direction:
            event, direction = event_direction.split('(')
            direction = direction[:-1]
        else:
            event = event_direction
            direction = None
        
        # check for punch in and out events and add to lists
        if event == 'in':
            punch_in_time = time
        elif event == 'out':
            punch_out_time = time
        else:
            continue
        
        # add punch in and out times to lists
        if punch_in_time and punch_out_time:
            punch_in_times.append(punch_in_time)
            punch_out_times.append(punch_out_time)
            attendance_dates.append(attendance_date)
            punch_in_time = None
            punch_out_time = None

# create pandas DataFrame
punch_df = pd.DataFrame({'attendance_date': attendance_dates, 'punch_in': punch_in_times, 'punch_out': punch_out_times})

# group by attendance_date
grouped_df = punch_df.groupby('attendance_date').apply(lambda x: x.reset_index(drop=True))

# merge grouped_df with original DataFrame
merged_df = pd.concat([df, grouped_df], axis=1)

# display output
output = merged_df[['employee_id', 'attendance_log_id', 'attendance_date', 'punch_in', 'punch_out']]
output.tail(2000)
```

```python
import mysql.connector
import pandas as pd

# Connect to MySQL database
db = mysql.connector.connect(
    host="localhost",
    user="root",
    password="root",
    database="sys"
)

# Define MySQL query
query = "SELECT employee_id, attendance_log_id, attendance_date, punch_records FROM stg_attendancelogs WHERE employee_id = '2544' AND attendance_date = '2022-08-22'  ORDER BY 1,2,3,4"

# Execute MySQL query and read results into pandas DataFrame
df = pd.read_sql_query(query, db)

# Create empty lists for punch in and out times and attendance dates
punch_in_times = []
punch_out_times = []
attendance_dates = []

# Loop through punch records and extract punch in and out times and attendance dates
for punch_record in df['punch_records'].unique():
    # Split punch record into individual events
    events = punch_record.split(',')
    
    # Initialize variables for tracking punch in and out times
    punch_in_time = None
    punch_out_time = None
    
    # Loop through events and extract punch in and out times
    for e in events:
        # Split event into time and event/direction
        split_e = e.split(':')
        if len(split_e) < 2:
            continue
        # Extract time and event information
        time = ':'.join(split_e[:2])
        event_direction = split_e[2]
        if '(' in event_direction and ')' in event_direction:
            event, direction = event_direction.split('(')
            direction = direction[:-1]
        else:
            event = event_direction
            direction = None
        
        # Check for punch in and out events and add to lists
        if event == 'in':
            punch_in_time = time
        elif event == 'out':
            punch_out_time = time
        else:
            continue
        
        # Add punch in and out times and attendance date to lists
        if punch_in_time and punch_out_time:
            punch_in_times.append(punch_in_time)
            punch_out_times.append(punch_out_time)
            attendance_dates.append(df.loc[df['punch_records'] == punch_record, 'attendance_date'].iloc[0])
            punch_in_time = None
            punch_out_time = None

# Create pandas DataFrame
punch_df = pd.DataFrame({'attendance_date': attendance_dates, 'punch_in': punch_in_times, 'punch_out': punch_out_times})

# Merge punch_df with original DataFrame
merged_df = pd.merge(df, punch_df, on='attendance_date', how='inner')

# Select only relevant columns and remove duplicates
output = merged_df[['employee_id', 'attendance_date', 'punch_in', 'punch_out']].drop_duplicates()

# Save output DataFrame to MySQL database
cursor = db.cursor()
for i, row in output.iterrows():
    cursor.execute("INSERT INTO attendance_records (employee_id, attendance_date, punch_in, punch_out) VALUES (%s, %s, %s, %s)", (row['employee_id'], row['attendance_date'], row['punch_in'], row['punch_out']))
db.commit()

# Close database connection
db.close()

# Display output
print(output)
```

```python
import mysql.connector
import pandas as pd


def tableOutput(dataArray):
    punches = dataArray[0]
    attendancedate = dataArray[1]
    employeeid = dataArray[2]
    
    employee_id=[]
    attendance_dates=[]
    insArray=[]
    outsArray=[]
    
    arrayOfinOuts=punches.split(",")
    
    allsorteddata=[]
    
    for i in range(len(arrayOfinOuts)):
        if arrayOfinOuts.index(arrayOfinOuts[i])==i:
            allsorteddata.append(arrayOfinOuts[i])
            
    arrayOfinOuts=allsorteddata
    
    inCons=0
    outsCons=0
    
    for inouts in arrayOfinOuts:
        if "in" in inouts and inCons==0:
            insArray.append(inouts)
            inCons+=1
            outsCons=0
        elif "in" in inouts and inCons>0:
            insArray.append(inouts)
            outsArray.append("N:A")
        elif "out" in inouts and outsCons==0:
            outsArray.append(inouts)
            outsCons+=1
            inCons=0
        elif "out" in inouts and outsCons>0:
            outsArray.append(inouts)
            insArray.append("N:A")

    updatedins=[]
    updatedouts=[]

    for i in insArray:
        a=i.split(":")
        updatedins.append(':'.join(a[:2]))

    for i in outsArray:
        a=i.split(":")
        updatedouts.append(':'.join(a[:2]))

    insArray=updatedins
    outsArray=updatedouts

    grtLen=0

    if len(insArray)>=len(outsArray):
        grtLen=len(insArray)
    else:
        grtLen=len(outsArray)

    for i in range(grtLen):
        attendance_dates.append(attendancedate)
        employee_id.append(employeeid)
        
    if len(insArray)==len(outsArray):
        pass
    elif len(insArray)>len(outsArray):
        outsArray.append("N:A")
    else:
        insArray.append("N:A")

    punch_in_times=insArray
    punch_out_times=outsArray

    # Create pandas DataFrame
    punch_df = pd.DataFrame({'employee_id' : employee_id, 'attendance_date': attendance_dates, 'punch_in': punch_in_times, 'punch_out': punch_out_times})

    # Convert employee_id to int type
    punch_df['employee_id'] = punch_df['employee_id'].astype(int)

    # Display output
    return(punch_df)

# Connect to MySQL database
db = mysql.connector.connect(
    host="localhost",
    user="root",
    password="root",
    database="sys"
)

# Define MySQL query
query = "SELECT employee_id, attendance_log_id, attendance_date, punch_records FROM stg_attendancelogs"

# Execute MySQL query and read results into pandas DataFrame
df = pd.read_sql_query(query, db)

# Create empty lists for punch in and out times and attendance dates


allPunches=[]
allDates=[]
allEmployes=[]


for punchrecs in df['punch_records']:
    allPunches.append(punchrecs)
for attendanceDate in df['attendance_date']:
    allDates.append(attendanceDate)
for employeeId in df['employee_id']:
    allEmployes.append(employeeId)

allData=[]

for i in range(len(allPunches)):
    if allPunches[i]!="":
        allData.append([allPunches[i], allDates[i], allEmployes[i]])
        
data = {'employee_id':  [""],
        'attendance_date': [""],
        'punch_in': [""],
        'punch_out': [""]
        }
allDataMerged=pd.DataFrame(data)
        
for j in allData:
    allDataMerged= pd.concat([allDataMerged, tableOutput(j)])
    
allDataMerged.head(30)
```
